```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_data = sklearn.datasets.load_breast_cancer()
df = pd.DataFrame(breast_cancer_data.data, columns= breast_cancer_data.feature_names)
df.head()
```

```python
df['label'] = breast_cancer_data.target
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.label.value_counts()
```

```python
df.groupby('label').mean()
```

```python
X = df.drop(columns='label', axis=1)
Y = df['label']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=100)
```

```python
# Standardize the data

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.fit_transform(X_test)

X_train = X_train_std
X_test = X_test_std
```

```python
# Artifical Neural Network

import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# ANN layers

model = keras.Sequential([
    # Input Layer
    keras.layers.Flatten(input_shape=(30,)),
    # Hiiden Layer
    keras.layers.Dense(20, activation='relu'),
    # Output Layer
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model_loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='upper right')
```

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print(loss, accuracy)
```

```python
# model.predict gives probability values of each class of the data

predicted = model.predict(X_test)

# conver prediction probability to class labels

labels = [np.argmax(i) for i in predicted]

print(labels)
```
